```python
# import packages
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib.pyplot as plt
import matplotlib
plt.style.use('ggplot')
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
# data load
df = pd.read_csv("income_per_person_gdppercapita_ppp_inflation_adjusted.csv")
df.shape
```

```python
df.describe()
```

```python
# % of missing.
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# filter data upon 4 countries
countries_to_keep = ['United States', 'China', 'India', 'Ethiopia']
columns = ['country','2000','2001','2002','2003','2004','2005','2006','2007','2008','2009','2010','2011','2012','2013','2014','2015','2016','2017','2018','2019','2020',]

df_d_income = df[df['country'].isin(countries_to_keep)][columns]
df_d_income
```

```python
# row to column
df_converted = df_d_income.melt(id_vars=["country"], 
        var_name="year", 
        value_name="income")
df_converted
```

```python
# convert 2.9 K(object) to 2900 (int)
repl_dict = {'[kK]': '*1e3', '[mM]': '*1e6', '[bB]': '*1e9', }
df_converted['income'].replace(repl_dict, regex=True).map(pd.eval)

sns.barplot(x='country', y=df_converted['income'].replace(repl_dict, regex=True).map(pd.eval), data=df_converted, hue='year')
```

```python
sns.lineplot(x = "year", y=df_converted['income'].replace(repl_dict, regex=True).map(pd.eval), hue='country', data = df_converted)
plt.show()
```
